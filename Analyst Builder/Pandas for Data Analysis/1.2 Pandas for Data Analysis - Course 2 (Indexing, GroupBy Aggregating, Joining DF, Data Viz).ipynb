## 1.0 Indexing

### 1.1 Indexing Introduction
Indexing is the process of accessing and retrieving data in a DataFrame or Series using labels or positions.

### 1.6 Reshaping and Pivoting Indexes
- **Reshaping**: Rearranging data structure from wide to long format or vice versa.
- **Forms of Reshaping**:
  - **Stacking**: Moves columns into the row index, creating a multi-level index (wide → long).
  - **Unstacking**: Moves levels of the row index into columns (long → wide)
- **Pivoting**: Converting unique values from one column into new columns, reorganizing data


```python
# Setup
import pandas as pd

# Importing df
df = pd.read_excel('/content/tech_price.xlsx')
df # Original raw output
```

```python
# Updating df in read_excel

df = pd.read_excel('/content/tech_price.xlsx', header = [0,1], index_col = [0,1])
df # Updated output

# `header=[0, 1]`: Specifies that the first two rows in the file are multi-level column headers.
# `index_col=[0, 1]`: Sets the first two columns as a multi-level row index.
```

```python
# 1.1.1 Reshaping Indexes: Stacking
df_stack = df.stack()
df_stack

# Stacks columns into the row index, converting from wide to long format
```

```python
# 1.1.2 Reshaping Indexes: Unstacking
df_stack.unstack()

# Unstacks row index levels into columns, converting from long to wide format
```

## 2.0 Group By and Aggregating

### 2.1 Group By and Aggregations Introduction
**Grouping**:
- Divides a DataFrame into groups based on column values
- **Key Components**:
  - **Grouping Column(s)**: Defines the groups
  - **GroupBy Object**: Intermediate object for applying aggregations

**Aggregations**:
- Calculates a single value for each group
- **Common Functions**:
  - `Sum`: Total of group values
  - `Mean`: Average of group values
  - `Count`: Non-null value count
  - `Min/Max`: Smallest and largest group values
  - `Custom`: Apply user-defined functions

### 2.2 Basics of Group By

```python
# Setup
import pandas as pd

df = pd.read_csv('/content/banking.csv')
df.head(5)
```

```python
# 1.1.1 Group By Part 1
df_marital_grouped = df.groupby('marital')['outstanding_balance'].sum()
df_marital_grouped

# Groups the DataFrame by the 'marital' column and computes the total outstanding balance for each marital status.
```

```python
# 1.1.2 Group By Part 2
df_job_grouped = df.groupby('job')['age'].mean()
df_job_grouped

# Groups the DataFrame by the 'job' column and computes the average age for each job category.
```

```python
# 1.1.3 Group By Part 3
df_job_grouped = df.groupby('job').describe()
df_job_grouped

# Groups the DataFrame by the 'job' column and generates descriptive statistics (count, mean, std, min, max, etc.) for each group.
```

### 2.3 Aggregations with Group By

```python
# Setup
import pandas as pd

df = pd.read_csv('/content/banking.csv')
df.head(5)
```

```python
# 1.1.1 Aggregations with Group By via min()
df.groupby('education')['outstanding_balance'].min()

# Groups by 'education' and computes the minimum outstanding balance for each group.
```

```python
# 1.1.2 Aggregations with Group By via max()
df.groupby('education')['outstanding_balance'].max()

# Groups by 'education' and computes the maximum outstanding balance for each group.
```

```python
# 1.1.3 Aggregations with Group By via sum()
df.groupby('education')['outstanding_balance'].sum().sort_values(ascending = False)

# Groups by 'education', sums the outstanding balances, and sorts in descending order.
```

```python
# 1.1.4 Aggregations with Group By via count()
df.groupby('education')['outstanding_balance'].count().sort_values(ascending = False)

# Groups by 'education', counts non-null outstanding balance entries, and sorts in descending order.
```

```python
# 1.1.5 Aggregations with Group By via median()
df.groupby('education')['outstanding_balance'].median().sort_values(ascending = False)

# Groups by 'education', computes the median outstanding balance, and sorts in descending order.
```

```python
# 1.1.6 Aggregations with Group By via sum() & count()
df.groupby('education').agg({'outstanding_balance': ['sum','count']})

# Groups by 'education' and applies multiple aggregations (sum and count) on 'outstanding_balance'.
```

```python
# 1.1.7 Aggregations with Group By via agg(), sum(), count()
output = df.groupby('education').agg({'outstanding_balance': ['sum','count']})

output['average'] = output['outstanding_balance']['sum']/output['outstanding_balance']['count']
output.sort_values(by = ('average'), ascending = False)

# Groups by 'education' and computes the sum and count of outstanding balances.
# Calculates the average outstanding balance for each group.
```

```python
# 1.1.8 Aggregations with Group By via agg()
df.groupby('education').agg(sum_balance = ('outstanding_balance','sum'), count_balance = ('outstanding_balance','count'))

# Groups by 'education' and uses the `agg` method to name the aggregated columns (sum and count of outstanding balances).
```

### 2.5 Transform with Group By
- Like Windows Function in SQL

```python
# Setup
import pandas as pd

df = pd.read_csv('/content/banking.csv')
df.head(5)
```

```python
# 1.1.1 Transform with Group By
df_banking = df[:500]

df_banking['average_balance'] = df_banking.groupby('job')['outstanding_balance'].transform('mean')
df_banking

# Creates a subset of the DataFrame containing the first 500 rows.
# Groups by 'job' and computes the mean outstanding balance for each job group.
# Uses `transform` to broadcast the mean value to each row in the 'job' group as a new column 'average_balance'.
```

```python
# 1.1.2 Transform with Group By
df_banking['above_below_balance'] = df_banking.groupby('job')['outstanding_balance'].transform(lambda x: x - x.mean())
df_banking.head(5)

# Groups by 'job' and calculates the difference between each row's outstanding balance and the group mean.
# Adds a new column 'above_below_balance' to indicate how much each row is above or below the group average.
```

## 3.0 Joining DataFrames

### 3.1 Joins Introduction
- Combining two or more dataframes together into one
- Ways to join dataframes in Pandas: Concat, Merge, Join
- `concat()`:
  - Concatenates two or more Pandas objects along a particular axis
  - Like union in SQL
- `merge()`:
  - Merge dataframes on a key or multiple keys by passing a list of column names
  - **Main Parameters**: 'on' and 'how'
- `join()`:
  - Used to combine columns of two potentially differently indexed dataframes into a single result
  - Equivalent to merge but operates on indices instead

### 3.2 Basics of Merge and Join
- Merge uses columns and Join uses index

```python
# Setup
import pandas as pd
from IPython.display import display_html

data1 = {
    'employee_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Dwight Schrute', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson'],
    'job_title': ['Regional Manager', 'Sales Representative', 'Receptionist', 'Assistant to the Regional Manager', 'Head of Accounting', 'Accountant', 'Accountant', 'Sales Representative'],
    'salary': [100000, 60000, 55000, 65000, 60000, 55000, 55000, 60000]
}

df1 = pd.DataFrame(data1)

data2 = {
    'employee_id': [1, 2, 3, 5, 6, 7, 8, 9],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson', 'Phyllis Vance'],
    'department': ['Management', 'Sales', 'Reception', 'Accounting', 'Accounting', 'Accounting', 'Sales', 'Sales']
}

df2 = pd.DataFrame(data2)

# Displaying both DataFrames side by side
space = "<div style='display:inline-block; width: 50px;'></div>"

display_html(
    df1.style.set_table_attributes("style='display:inline; margin-right: 50px;'")._repr_html_() +
    space +
    df2.style.set_table_attributes("style='display:inline'")._repr_html_(),
    raw=True
)
```

```python
# 1.1.1 Basics of Merge
df1.merge(df2) # Automatically joined inner and by employee_id/name

# Performs an inner join on matching columns by default (common columns like 'employee_id' or 'name').
```

```python
# 1.1.2 Basics of Merge: Specifying the how
df1.merge(df2, how = 'left') # Left join
```

```python
# 1.1.3 Basics of Merge: Specifying the how
df1.merge(df2, how = 'right') # Right join
```

```python
# 1.1.4 Basics of Merge: Specifying the how
df1.merge(df2, how = 'outer') # Outer join
```

```python
# 1.1.5 Basics of Merge: Specifying the how
df1.merge(df2, how = 'outer', on = 'employee_id') # name_x and name_y: Why is there separate name columns?

# Outer join on 'employee_id'; creates separate columns 'name_x' and 'name_y' if both DataFrames have a 'name' column.

# Why separate name columns?
  # If both DataFrames have a column with the same name but different values, Pandas adds suffixes ('_x' for left DataFrame, '_y' for right) to differentiate them.
```

```python
# Setup

data1_ = {
    'emp_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Dwight Schrute', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson'],
    'job_title': ['Regional Manager', 'Sales Representative', 'Receptionist', 'Assistant to the Regional Manager', 'Head of Accounting', 'Accountant', 'Accountant', 'Sales Representative'],
    'salary': [100000, 60000, 55000, 65000, 60000, 55000, 55000, 60000]
}

df1_ = pd.DataFrame(data1_)

data2_ = {
    'employee_id': [1, 2, 3, 5, 6, 7, 8, 9],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson', 'Phyllis Vance'],
    'department': ['Management', 'Sales', 'Reception', 'Accounting', 'Accounting', 'Accounting', 'Sales', 'Sales']
}

df2_ = pd.DataFrame(data2_)

# Displaying both DataFrames side by side
space = "<div style='display:inline-block; width: 50px;'></div>"

display_html(
    df1_.style.set_table_attributes("style='display:inline; margin-right: 50px;'")._repr_html_() +
    space +
    df2_.style.set_table_attributes("style='display:inline'")._repr_html_(),
    raw=True
)
```

```python
# 1.1.6 Basics of Merge: Specifying the how
df1_.merge(df2_, how = 'outer', left_on = 'emp_id', right_on = 'employee_id')

# Outer join on 'emp_id' (from df1_) and 'employee_id' (from df2_); ensures rows are matched correctly based on these columns.
```

```python
# 1.1.7 Basics of Join:
df1_.set_index('emp_id', inplace = True)
df2_.set_index('employee_id', inplace = True)

df1_.join(df2, how = 'inner', lsuffix = '_df1', rsuffix = '_df2')

# Sets 'emp_id' as the index for df1_.
# Sets 'employee_id' as the index for df2_.
# Performs an inner join on the index (emp_id/employee_id); adds suffixes '_df1' and '_df2' to avoid column name conflicts.
```

### 3.4 Merging Multiple DataFrames

```python
# Setup
import pandas as pd

data1 = {
    'employee_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Dwight Schrute', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson'],
    'job_title': ['Regional Manager', 'Sales Representative', 'Receptionist', 'Assistant to the Regional Manager', 'Head of Accounting', 'Accountant', 'Accountant', 'Sales Representative'],
    'salary': [100000, 60000, 55000, 65000, 60000, 55000, 55000, 60000]
}

demographics = pd.DataFrame(data1)

data2 = {
    'employee_id': [1, 2, 3, 5, 6, 7, 8, 9],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson', 'Phyllis Vance'],
    'department_id': [1, 2, 3, 4, 4, 4, 2, 2]
}

name_dep = pd.DataFrame(data2)


data3 = {
    'employee_id': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'manager_id': [None, 1, 1, 1, 1, 4, 4, 1, 1]
}


manager = pd.DataFrame(data3)
manager['manager_id'] = manager['manager_id'].astype('Int64')

data4 = {
    'department_id': [1, 2, 3, 4],
    'department': ['Management', 'Sales', 'Reception', 'Accounting']
}


department = pd.DataFrame(data4)

# Displaying both DataFrames side by side
space = "<div style='display:inline-block; width: 30px;'></div>"

display_html(
    demographics.style.set_table_attributes("style='display:inline; margin-right: 50px;'")._repr_html_() +
    name_dep.style.set_table_attributes("style='display:inline'")._repr_html_() + space +
    manager.style.set_table_attributes("style='display:inline'")._repr_html_() + space +
    department.style.set_table_attributes("style='display:inline'")._repr_html_() + space,
    raw=True
)
```

```python
# 1.1.1 Merging Multiple Names

pd.merge(demographics, name_dep, on = 'employee_id', how = 'inner').merge(department, on = 'department_id', how = 'inner').merge(manager, on = 'employee_id', how = 'inner')

# Merging multiple DataFrames sequentially to combine related data based on matching keys
```

### 3.5 Concatenation

```python
# Setup
import pandas as pd

data1 = {
    'employee_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Dwight Schrute', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson'],
    'job_title': ['Regional Manager', 'Sales Representative', 'Receptionist', 'Assistant to the Regional Manager', 'Head of Accounting', 'Accountant', 'Accountant', 'Sales Representative'],
    'salary': [100000, 60000, 55000, 65000, 60000, 55000, 55000, 60000]
}

df1 = pd.DataFrame(data1)

data2 = {
    'employee_id': [1, 2, 3, 5, 6, 7, 8, 9],
    'name': ['Michael Scott', 'Jim Halpert', 'Pam Beesly', 'Angela Martin', 'Kevin Malone', 'Oscar Martinez', 'Stanley Hudson', 'Phyllis Vance'],
    'department': ['Management', 'Sales', 'Reception', 'Accounting', 'Accounting', 'Accounting', 'Sales', 'Sales']
}

df2 = pd.DataFrame(data2)

display_html(
    df1.style.set_table_attributes("style='display:inline; margin-right: 50px;'")._repr_html_() +
    df2.style.set_table_attributes("style='display:inline'")._repr_html_(),
    raw=True
)
```

```python
# 1.1.1 Concatenation
pd.concat([df1, df2])

# Combines `df1` and `df2` along the default axis (rows) into one DataFrame, stacking them vertically.
```

```python
# 1.1.2 Concatenation: Main use case
pd.concat([df1, df2], axis = 0)

# Explicitly concatenates `df1` and `df2` along rows (axis=0). Useful for appending datasets.
```

```python
# 1.1.3 Concatenation: Not an actual use case
pd.concat([df1.set_index('employee_id'), df2.set_index('employee_id')], axis = 1)

# Combines `df1` and `df2` side-by-side (column-wise, axis=1) after setting `employee_id` as the index.
```

```python
# 1.1.4 Concatenation
pd.concat([df1,df2.rename(columns = {'department':'job_title'})], axis = 0)

# Renames the `department` column in `df2` to `job_title` before vertically concatenating it with `df1`.
```

```python
# 1.1.5 Concatenation
pd.concat([df1[['employee_id','name']],df2[['employee_id','name']]], axis = 0)

# Extracts `employee_id` and `name` columns from both DataFrames and concatenates them vertically.
```

```python
# 1.1.6 Concatenation
pd.concat([df1[['employee_id','name']],df2[['employee_id','name']]], axis = 0).drop_duplicates().reset_index().drop('index', axis = 1)

# Concatenates selected columns from `df1` and `df2`, removes duplicate rows, resets the index, and drops the old index column.
```

```python
# 1.1.7 Concatenation
pd.concat([df1, df2], axis = 0, ignore_index = True)

# Vertically concatenates `df1` and `df2` while ignoring original index values, resetting the index sequentially.
```

```python
# 1.1.8 Concatenation
pd.concat([df1, df2], axis = 0, keys = ['DF1','DF2'])

# Concatenates `df1` and `df2` vertically, adding a hierarchical index (`DF1`, `DF2`) to indicate source DataFrame.
```

```python
# Setup
import pandas as pd

df = pd.read_csv('/content/all_stocks_5yr.csv')
df.head(3)
```

```python
# 1.1.1 Visualizing data with Pandas: Line

df.plot(x = 'date', y = 'close', ylabel = 'Closing Price', xlabel = 'Date of Price', title = 'Stock Closing Prices')

# Creates a line plot showing closing prices over time with labeled axes and a title.
```

```python
# 1.1.2 Visualizing data with Pandas: Line

company_df = df[df['Name'] == 'AAL']
company_df.plot(x = 'date', y = 'close', ylabel = 'Closing Price', xlabel = 'Date of Price', title = 'AAL: Stock Closing Prices')

# Filters the DataFrame for a specific company ('AAL') and creates a line plot for its closing prices.
```

```python
# 1.1.3 Visualizing data with Pandas: Bar

company_bar = df.groupby('Name')['volume'].sum().reset_index()
company_bar = company_bar[company_bar['Name'].isin(['AAL','AAP','AAPL'])]

company_bar.plot(kind = 'bar', x = 'Name', y = 'volume', xlabel = 'Company', ylabel = 'Volume', title = 'Volume by Company')

# Aggregates volume data by company, filters specific companies, and visualizes the data as a bar chart.
```

```python
# 1.1.4 Visualizing data with Pandas: Scatter

company_df = df[df['Name'] == 'AAL']
company_df.plot(kind = 'scatter', x = 'open', y = 'close', xlabel = 'Open', ylabel = 'Close', title = 'Open vs Close')

# Creates a scatter plot to visualize the relationship between open and close prices for 'AAL'.
```

```python
# 1.1.5 Visualizing data with Pandas: Histogram

company_df = df[df['Name'] == 'AAL']
company_df['high'].plot.hist(bins = 25, xlabel = 'High')

# Generates a histogram of the 'high' prices for 'AAL', grouped into 25 bins.
```

```python
# 1.1.6 Visualizing data with Pandas: Pie

pie_df = df[df['Name'].isin(['AAL','GOOG','AAPL'])]
pie_df.groupby('Name')['volume'].sum().plot(kind = 'pie')

# Creates a pie chart showing the volume distribution for selected companies.
```

```python
# 1.1.7 Visualizing data with Pandas: Box

company_df = df[df['Name'] == 'AAL']
company_df[['open','high','low','close']].plot(kind = 'box')

# Generates a box plot to show the distribution of open, high, low, and close prices for 'AAL'.
```

```python
# 1.1.8 Visualizing data with Pandas: Box

company_df = df[df['Name'] == 'AAL']
company_df[['volume']].plot(kind = 'box')

# Creates a box plot to show the distribution of trading volumes for 'AAL'.
```

### 4.3 Introduction to Matplotlib

```python
# Setup

import matplotlib.pyplot as plt
```

```python
# 1.1.1 Matplotlib

x = [1, 2, 3]  # X-axis values
y = [5, 10, 20]  # Y-axis values

plt.plot(x, y, label='Label', color='green', marker='^', markersize=10, linestyle='--', markeredgecolor='red')
# Plot a single line with:
# - Label: 'Label'
# - Line color: Green
# - Marker: Triangle with red edges
# - Line style: Dashed

plt.title('Matplotlib Viz', fontdict={'fontsize': 30})  # Set the plot title with custom font size
plt.xlabel('X Axis')  # Set X-axis label
plt.ylabel('Y Axis')  # Set Y-axis label
plt.xticks([1, 2, 3])  # Customize X-axis tick values
plt.yticks([0, 5, 10, 15, 20])  # Customize Y-axis tick values
plt.legend()  # Add a legend to the plot
plt.show()  # Display the plot
```

```python
# 1.1.2 Matplotlib

x = [1, 2, 3]  # X-axis values
y = [5, 10, 20]  # Y-axis values for Line 1
z = [1, 5, 10]  # Y-axis values for Line 2

plt.figure(figsize=(20, 5))  # Set figure size to 20 (width) by 5 (height)

plt.plot(x, y, label='Label_Y', color='green', marker='^', markersize=10, linestyle='--', markeredgecolor='red')
# Plot Line 1 with:
# - Label: 'Label_Y'
# - Line color: Green
# - Marker: Triangle with red edges
# - Line style: Dashed

plt.plot(x, z, label='Label_Z', color='blue', marker='.', markersize=15, linestyle='--', markeredgecolor='red')
# Plot Line 2 with:
# - Label: 'Label_Z'
# - Line color: Blue
# - Marker: Dot with red edges
# - Line style: Dashed

plt.title('Matplotlib Viz', fontdict={'fontsize': 30})  # Set the plot title with custom font size
plt.xlabel('X Axis')  # Set X-axis label
plt.ylabel('Y Axis')  # Set Y-axis label
plt.xticks([1, 2, 3])  # Customize X-axis tick values
plt.yticks([0, 5, 10, 15, 20])  # Customize Y-axis tick values
plt.legend()  # Add a legend to describe each line
plt.show()  # Display the plot
```

### 4.4 Visualization with Seaborn

```python
# Setup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

housing = pd.read_csv('/content/real_estate.csv')
housing.head(3)
```

```python
# 1.1.1 Seaborn

housing_filtered = housing[housing['state'].isin(['Massachusetts', 'Puerto Rico', 'New York'])]
# Filters the data to include only rows where 'state' is one of the specified values.

sns.barplot(housing_filtered, x = 'state', y = 'price', estimator = 'mean')
# Creates a bar plot where the x-axis is 'state', the y-axis is the mean of 'price' for each state.

plt.plot()
```

```python
# 1.1.1 Panda

housing_filtered = housing[housing['state'].isin(['Massachusetts', 'Puerto Rico', 'New York'])]
# Filters the data to include only rows where 'state' matches one of the specified values.

housing_panda = housing_filtered.groupby('state')['price'].mean().reset_index()
# Groups the data by 'state', calculates the mean of 'price', and converts it into a new DataFrame.

housing_panda.plot(x = 'state', y = 'price', kind = 'bar')
# Creates a bar plot using Pandas with 'state' on the x-axis and the average price on the y-axis.
```

```python
# 1.1.3 Seaborn: set_style

housing_filtered = housing[housing['state'].isin(['Massachusetts', 'Puerto Rico', 'New York'])]
# Filters the data for the specified states.

sns.set_style('darkgrid')
# Sets the plot style to 'darkgrid', adding a grid background.

sns.set_palette('Reds')
# Sets the color palette for the plot to shades of red.

plt.title('Average State by Price')
# Adds a title to the plot.

sns.barplot(housing_filtered, x = 'state', y = 'price', estimator = 'mean')
# Creates a bar plot as before with the customized style.

plt.plot()
```

```python
# 1.1.4 Seaborn: regplot()

sns.set_style('darkgrid')
sns.set_palette('deep')
# Customizes the plot's appearance with a dark grid background and default deep color palette.

housing_filt = housing[housing['house_size'] < 10000]
# Filters the data to include houses smaller than 10,000 square feet.

sns.regplot(data = housing_filt, x = 'house_size', y = 'price', line_kws = {'color':'purple'})
# Creates a scatter plot of house size vs. price and fits a regression line (colored purple).

plt.ticklabel_format(style = 'plain', axis = 'y')
# Ensures y-axis tick labels are displayed in plain style (not scientific notation).

plt.show()
```

```python
# 1.1.5 Seaborn: pairplot()

iris = sns.load_dataset('iris')
# Loads the Iris dataset

sns.pairplot(iris, hue = 'species')
# Creates a grid of scatter plots for all numeric columns, color-coded by 'species'.

plt.show()
```

```python
# 1.1.6 Seaborn: stripplot()

iris = sns.load_dataset('iris')
# Loads the Iris dataset

sns.stripplot(data = iris, hue = 'species', x = 'species', y = 'sepal_length')
# Creates a strip plot showing data points for 'sepal_length' grouped by 'species', with color-coding for species.

plt.show()
```

```python
# 1.1.7 Seaborn: swarmplot()

iris = sns.load_dataset('iris')
# Loads the Iris dataset

sns.swarmplot(data = iris, hue = 'species', x = 'species', y = 'sepal_length')
# Creates a swarm plot that avoids overlapping points for 'sepal_length' grouped by 'species'.

plt.show()
```

```python
# 1.1.7 Seaborn: violinplot()

iris = sns.load_dataset('iris')
# Loads the Iris dataset

sns.violinplot(data = iris, hue = 'species', x = 'species', y = 'sepal_length')
# Creates a violin plot showing the distribution of sepal lengths for each species.

plt.show()
```

### 4.5 Advanced Data Visualization

```python
# Setup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

housing = pd.read_csv('/content/real_estate.csv')
smoking = pd.read_csv('/content/smoking.csv')
smoking.head(3)
```

```python
# 1.1.1 Advanced Data Viz

smokers = smoking[smoking['smoke'] == 'Yes'] # Filters the dataset to include only rows where 'smoke' is 'Yes', indicating smokers.
female_smokers = smoking[smoking['gender'] == 'Female'] # Filters the dataset to include only rows where 'gender' is 'Female'.
male_smokers = smoking[smoking['gender'] == 'Male'] # Filters the dataset to include only rows where 'gender' is 'Male'.

fig, axs = plt.subplots(1, 2) # Creates a figure with a 1x2 grid of subplots, stored in 'axs' as an array.

sns.histplot(female_smokers['age'], bins=25, ax=axs[0]) # Creates a histogram of the 'age' column for female smokers with 25 bins, plotted on the first subplot.
axs[0].set_title('Age Distribution: Female Smokers') # Sets the title for the first subplot.

sns.histplot(male_smokers['age'], bins=25, ax=axs[1]) # Creates a histogram of the 'age' column for male smokers with 25 bins, plotted on the second subplot.
axs[1].set_title('Age Distribution: Male Smokers') # Sets the title for the second subplot.

plt.tight_layout() # Adjusts the spacing between subplots to ensure they don't overlap.
plt.show()
```

```python
# 1.1.2 Advanced Data Viz: Correlation
housing_filt = housing.drop('zip_code', axis=1)  # Removes 'zip_code' column from the DataFrame as it is not relevant for correlation.

housing_float = housing_filt.select_dtypes(include=['float64'])  # Filters only columns with 'float64' data type for correlation analysis.
housing_float.corr()  # Calculates the Pearson correlation matrix for the filtered DataFrame.
```

```python
# 1.1.3 Advanced Data Viz: Correlation Heatmap
housing_filt = housing.drop('zip_code', axis=1)  # Drops 'zip_code' column again (if not already dropped).

housing_float = housing_filt.select_dtypes(include=['float64'])  # Filters columns with 'float64' data type.
correlation = housing_float.corr()  # Calculates the Pearson correlation matrix.

sns.heatmap(correlation, annot=True, cmap='coolwarm')  # Creates a heatmap to visualize correlations with annotations and a color gradient.
```

```python
# 1.1.4 Advanced Data Viz: Different Correlations

pearson_correlation = housing_float.corr(method='pearson')  # Calculates the Pearson correlation matrix.
print(pearson_correlation, '\n')  # Prints the Pearson correlation matrix.

spearman_correlation = housing_float.corr(method='spearman')  # Calculates the Spearman correlation matrix.
print(spearman_correlation, '\n')  # Prints the Spearman correlation matrix.

kendall_correlation = housing_float.corr(method='kendall')  # Calculates the Kendall Tau correlation matrix.
print(kendall_correlation)  # Prints the Kendall correlation matrix.
```

```python
# 1.1.5 Advanced Data Viz: Pearson
sns.heatmap(pearson_correlation, annot=True, cmap='coolwarm')  # Heatmap for the Pearson correlation matrix.
```

```python
# 1.1.6 Advanced Data Viz: Spearman
sns.heatmap(spearman_correlation, annot=True, cmap='coolwarm')  # Heatmap for the Spearman correlation matrix.
```

```python
# 1.1.7 Advanced Data Viz: Kendall
sns.heatmap(kendall_correlation, annot=True, cmap='coolwarm')  # Heatmap for the Kendall Tau correlation matrix.
```

```python
# 1.1.8 Fig Size

fig, axs = plt.subplots(3, 1, figsize=(7, 10))  # Creates a figure with 3 vertically stacked subplots and specified figure size.

sns.heatmap(pearson_correlation, annot=True, cmap='coolwarm', ax=axs[0])  # Heatmap for Pearson correlation matrix in the first subplot.
axs[0].set_title('Pearson Correlation')  # Sets title for the first subplot.

sns.heatmap(spearman_correlation, annot=True, cmap='coolwarm', ax=axs[1])  # Heatmap for Spearman correlation matrix in the second subplot.
axs[1].set_title('Spearman Correlation')  # Sets title for the second subplot.

sns.heatmap(kendall_correlation, annot=True, cmap='coolwarm', ax=axs[2])  # Heatmap for Kendall correlation matrix in the third subplot.
axs[2].set_title('Kendall Correlation')  # Sets title for the third subplot.

plt.tight_layout()  # Adjusts spacing between subplots to prevent overlap.
plt.show()  # Displays the figure with all heatmaps.
```

**Pearson Correlation (Linear Correlation)**
- **What it Measures**: The linear relationship between two continuous variables.
- **Key Features**:
  - Assumes a linear relationship between variables.
  - Sensitive to outliers, as extreme values can greatly affect the result.
  - Values range from -1 (perfect negative linear) to +1 (perfect positive linear).
- **Use Case**: When variables have a linear relationship and data is normally distributed.

**Spearman Correlation (Rank-Based)**
- **What it Measures**: The monotonic relationship between two variables (whether they move in the same or opposite direction).
- **Key Features**:
  - Based on ranked data, ignoring actual values.
  - Less sensitive to outliers than Pearson since it ranks the data.
  - Suitable for non-linear but monotonic relationships.
- **Use Case**: When variables may not have a linear relationship or are ordinal.

**Kendall Correlation (Rank-Based)**
- **What it Measures**: The strength of association between two variables by comparing the order or concordance of pairs.
- **Key Features**:
  - Measures the proportion of concordant and discordant pairs.
  - More robust against outliers compared to both Pearson and Spearman.
  - Less sensitive to small sample sizes.
- **Use Case**: When assessing ordinal data or pairwise concordance
