## 2.0 Regular Expression

### 2.1 Regular Expression Introduction:
- Used to match strings of text (characters, patterns of characters, words)
- Known as `Regex` or `Regexp`
- Can help users replace, remove, or extract specific patterns within strings

### 2.2 Re Module + Regex Methods:
- re module provides functions and methods for working with regular expressions
- **Regular expression methods**:
  - `re.search`: Finds pattern anywhere in string, returns match object.
  - `re.match`: Checks if pattern matches at start of string, returns match object.
  - `re.fullmatch`: Matches whole string to pattern, returns match object.
  - `re.findall`: Returns all pattern matches in string as a list.
  - `re.finditer`: Returns iterator of match objects for all pattern matches.
  - `re.split`: Splits string by pattern, returns list.
  - `re.sub`: Replaces pattern matches with given string.
  - `re.subn`: Same as re.sub, but also returns number of replacements.
  - `re.compile`: Compiles pattern for reuse.
  - `re.escape`: Escapes special characters in string for regex.



```python
# Setup

import re
```

```python
#1.1.1 Common regular expression examples

quote = 'I saved a life. My own. Am I a hero? I really cannot say, but yes. - Michael Scott'

print(re.search('I', quote))  # Find first occurrence of 'I' and return match object
print(re.findall('I', quote))  # Find all occurrences of 'I' and return as a list
print(re.split('I', quote))  # Split string at each 'I', return list of parts
print(re.split('\.', quote))  # Split string at each period '.', return list of parts
print(re.split('\.', quote, maxsplit=1))  # Split at first period only, return two parts
print(re.sub('I', 'You', quote))  # Replace all 'I' with 'You'
print(re.sub('I', 'You', quote, maxsplit=1))  # Replace first 'I' with 'You' only
```

### 2.3 Regex Character Classes:
- `[]`: Defines a character class (matches any single character within the brackets).
- `\`: Used for escaping Metacharacters
- `.`: Matches any character except a newline
- `^`: Anchors the regex at the start of the string
- `$`: Anchors the regex at the end of the string
- `*`: Matches 0 or more occurrences of the preceding character
- `+`: Matches 1 or more occurrences of the preceding character
- `?`: Matches 0 or 1 occurrence of the preceding character
- `{}`: Exactly the specified number of occurrences
- `|`: Acts like a logical OR. A|B matches either A or B
- `()`: Capture and group

```python
# Setup

import re
```

```python
# 1.1.1 Regex character '[]' example

string1 = 'I enjoy fajitas in the winter'
print(re.findall('[a-eA-E]', string1))  # Find all letters a-e (case-insensitive) in the string

string2 = 'I have 123,456 cats'
print(re.findall('[1-3]', string2))  # Find all digits 1-3 in the string
```

```python
# 1.1.2 Regex character '\' example

string2 = 'I have 123,456 cats' # Find all digits in the string
print(re.findall('\d', string2))
```

```python
# 1.1.3 Regex character '.' example
string3 = 'hello hero helping helps'

print(re.findall('.', string3))  # Find every character
print(re.findall('he.', string3))  # Find 'he' followed by any character
print(re.findall('he.o', string3))  # Find 'he', any character, then 'o'
print(re.findall('he..o', string3))  # Find 'he', any two characters, then 'o'
```

```python
# 1.1.4 Regex character '^', '$' example

string4 = 'sunflowers, sundrop, santa clause, and mushrooms are lovely in the spring'

print(re.findall('^sun', string4))  # Check if string starts with 'sun'
print(re.findall('ing$', string4))  # Check if string ends with 'ing'
```

```python
# 1.1.5 Regex character '*', '+', '?' example

string5 = 'This Thing called a Thimble is Thick'
print(re.findall('Thi.*s', string5))  # Match 'Thi', followed by any characters, ending in 's'

string6 = 'Helping some is kind'
print(re.findall('H.*ing', string6))  # Match 'H', any characters, ending in 'ing'
print(re.findall('H.+ing', string6))  # Match 'H', at least one character, ending in 'ing'

string7 = 'Hing someone is kind'
print(re.findall('H.*ing', string7))  # Match 'H', any characters, ending in 'ing' (zero or more characters allowed)
print(re.findall('H.+ing', string7))  # Match 'H', at least one character, ending in 'ing'

string8 = 'Hing, Helping someone is kind. I like Huing'
print(re.findall('H.?ing', string8))  # Match 'H', zero or one character, ending in 'ing'
```

```python
# 1.1.6 Regex character '{}' example

string9 = 'Where are Whinie While I look around'

print(re.findall('Wh.{2}e', string9))  # Match 'Wh', any two characters, then 'e'
print(re.findall('Wh.{3}e', string9))  # Match 'Wh', any three characters, then 'e'
print(re.findall('Wh...e', string9))  # Match 'Wh', exactly three characters, then 'e'
```

```python
# 1.1.7 Regex character '|' example

string10 = 'Its a beautiful day in the city with all the beautiful lights. Its lovely.'

print(re.findall('beautiful|lovely', string10))  # Match either 'beautiful' or 'lovely'
```

### 2.4 Regex MetaCharacters:
- `[abc]`: Matches any single character that is either a, b, or c.
- `[^abc]`: Matches any single character that is not a, b, or c. The ^ inside the square brackets negates the character class.
- `[a-z]`: Matches any single lowercase letter from a to z.
- `[A-Z]`: Matches any single uppercase letter from A to Z.
- `[0-9]`: Matches any single digit from 0 to 9.
- `[a-zA-Z]`: Matches any single letter, either lowercase or uppercase.
- `[\d]`: Matches any single digit. It is equivalent to [0-9].
- `[\D]`: Matches any single non-digit character. It is equivalent to [^0-9].
- `[\w]`: Matches any single alphanumeric character (letters, digits, or underscore). It is equivalent to [a-zA-Z0-9_].
- `[\W]`: Matches any single non-alphanumeric character. It is equivalent to [^a-zA-Z0-9_].
- `[\s]`: Matches any whitespace character (space, tab, newline, etc.).
- `[\S]`: Matches any non-whitespace character.
- `[\b]`: Matches when there is a word boundary present (spaces mostly).
- `[\B]`: Matches there is not a word boundary present (characters).
- `[\Z]`: Matches at the end of a string

```python
# 1.1.1 Regex metacharacter example

jc_quote1 = 'My name is Mr. Choi. My phone number is 740-517-6485. My email is josephchoi.com'

print(re.findall('\d', jc_quote1))  # Find all digits
print(re.findall('\w', jc_quote1))  # Find all alphanumeric characters (letters, digits, underscores)
print(re.findall('\W', jc_quote1))  # Find all non-alphanumeric characters (spaces, punctuation)
print(re.findall('\A[A-Z]', jc_quote1))  # Check if string starts with an uppercase letter
print(re.findall('[a-zA-Z]\Z', jc_quote1))  # Check if string ends with a letter
```

```python
# 1.1.2 Regex metacharacter example

jc_quote2 = 'I like Python is PythonLove a word?'

print(re.findall(r'\bPython', jc_quote2))  # Find 'Python' as a standalone word at word boundaries
print(re.findall(r'\bPython\b', jc_quote2))  # Match 'Python' as a complete word with boundaries on both sides
print(re.findall(r'\Bython', jc_quote2))  # Find 'ython' occurring within a word (no boundary at start)
print(re.findall(r'\Bython\B', jc_quote2))  # Find 'ython' within a word with no boundaries at either end
```

```python
# 1.1.3 Regex metacharacter example

jc_quote3 = 'There are 365 days in a year. 8760 hours. 525600 minutes. Only 24 days until Christmas. 1 Day at a time'

print(re.findall(r'\d', jc_quote3))  # Find all digits individually
print(re.findall(r'\d{3}', jc_quote3))  # Find groups of exactly 3 consecutive digits
print(re.findall(r'\d{3,4}', jc_quote3))  # Find groups of 3 or 4 consecutive digits
print(re.findall(r'\d+', jc_quote3))  # Find all continuous digit sequences
print(len(re.findall(r'\d+', jc_quote3)))  # Count the number of digit sequences found
```

### 2.5 Match Object Methods:
- `re.match()`, `re.search()`, and `re.findall()` functions return a match object that contains information about the pattern match
- The match object provides various methods to access and manipulate the matched content
- **Match Objects**:
  - `group()`: Returns the matched string.
  - `start()`: Returns the starting position of the match.
  - `end()`: Returns the ending position of the match.
  - `span()`: Returns a tuple containing the starting and ending positions of the match.
  - `groups()`: Returns a tuple containing all the captured groups in the match.
  - `group(index)`: Returns the captured group at the specified index.
  - `groupdict()`: Returns a dictionary containing named captured groups, where the keys are the group names.
  - `expand(template)`: Returns the matched string with any backreferences replaced using the specified template.


```python
# 1.1.1 Match object method example

jc_string1 = "It's a BEAUTIFUL day in the city. It's a 75 degrees and sunny!"
print(re.search(r'(\b[A-Z]+\b).*(\b\d+\b)', jc_string1))  # Search for uppercase word followed by a number

result = re.search(r'(\b[A-Z]+\b).*(\b\d+\b)', jc_string1)
print('\nprint1:', result.group())  # Full matched string
print('print2:', result.groups())  # Tuple of matched groups (uppercase word, number)
print('print3:', result.group(1))  # First matched group (uppercase word)
print('print4:', result.group(2))  # Second matched group (number)
print('print5:', result.start())  # Start index of the full match
print('print6:', result.end())  # End index of the full match
print('print7:', result.span())  # Tuple with start and end index of the full match
```

### 2.6 Regex Modifiers:
- Modifiers (also known as flags) are used to modify the behavior of the pattern matching
- They are added as optional parameters to the regular expression functions or as inline modifiers within the pattern
- **Regex Modifiers**:
  - `re.IGNORECASE or re.I`: Ignores case sensitivity when matching characters. For example, /abc/i matches "abc", "AbC", "ABC", etc.
  - `re.MULTILINE or re.M`: Enables multi-line mode. Changes the behavior of ^ and $ to match the start and end of each line in addition to the start and end of the string.
  - `re.DOTALL or re.S`: Makes the . metacharacter match any character, including newline characters (\n).
  - `re.VERBOSE or re.X`: Allows the use of whitespace and comments within the regular expression pattern for better readability. Whitespace and comments are ignored unless they are within square brackets or escaped.
  - `re.ASCII or re.A`: Restricts character interpretation to the ASCII character set. Overrides the use of Unicode for character classes like \w, \W, \b, etc.
  - `re.UNICODE or re.U`: Enables Unicode matching. By default, \w, \W, \b, etc., use Unicode property escapes to match word characters and word boundaries.
  - `re.DEBUG`: Prints debugging information about the pattern compilation.

```python
# 1.1.1 Regex modifiers example

jc_quote1 = 'My name is Mr. Choi. My phone number is 740-517-6485. My email is josephchoi.com'

print(re.findall(r'\b[a-z]+\b', jc_quote1))  # Find lowercase words (case-sensitive)
print(re.findall(r'\b[a-z]+\b', jc_quote1, re.I))  # Find words ignoring case (re.I for case-insensitive)
```

```python
# 1.1.2 Regex modifiers example

target_string = 'Joseph likes the number 11\nAlexies likes the number 16'

print(re.findall(r'^\w{4}', target_string))  # Match 4-character word at the start of the string
print(re.findall(r'^\w{4}', target_string, re.M))  # Match 4-character word at the start of each line (re.M for multiline)
print(re.findall(r'\d{2}$', target_string))  # Match 2-digit number at the end of the string
print(re.findall(r'\d{2}$', target_string, re.M))  # Match 2-digit number at the end of each line (re.M for multiline)
```

```python
# 1.1.3 Regex modifiers example

string_101 = 'I like ice cream.\nI like pie as well!'
print('print1:', re.findall(r'.+', string_101))  # Match any characters except newlines for each line separately
print('print2:', re.findall(r'.+', string_101, re.S))  # Match any characters including newlines (re.S for single line mode)

result1 = re.search(r'.+', string_101)
result2 = re.search(r'.+', string_101, re.S)
print('\nprint3:', result1.group())  # Match first line only (without newline)
print('print4:', result2.group())  # Match entire text including newline with re.S
```

### 2.7 Regex Use Cases:

```python
# 1.1.1 Regex use cases example

random_text = '''
My name is Mr. Joseph. My phone number is 546-515-8489. My email is josephchoi@gmail.com
My name is Mr. Joe. My phone number is 546-515-8489. My email is joechoi@hotmail.com
My name is Mr. Joey. My phone number is 546-515-8489. My email is joeychoi@aol.com
'''

print(re.findall(r'@([a-zA-Z]+)', random_text))  # Find email domain names only (e.g., 'gmail', 'hotmail', etc.)
print(re.findall(r'@([\w\.+]+)', random_text))  # Find domain part of emails with special characters
print(re.findall(r'[\w\.+]+@[\w\.+]+', random_text), '\n')  # Find full email addresses

print(re.findall(r'\d{3}-\d{3}-\d{4}', random_text))  # Find phone numbers in '123-456-7890' format
print(re.findall(r'\ (\d{3})', random_text))  # Find and capture area codes as three-digit groups
```

```python
# 1.1.2 Regex use cases example

email_list = ['josephchoi@gmail.com','joechoi@hotmail.com','joeychoi@aol.com']

for email in email_list:
  print(re.findall(r'[\w\.+]+@[\w\.+]+', email), '\n')  # Print each email address using regex
```
